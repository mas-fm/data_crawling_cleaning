# import with loop

```python
import pandas as pd
from pathlib import Path, PurePath
path = r'D:\User\Desktop\thesis\data\tarazname\balance sheet\{}.xlsx'
df_list = []
for i in range(1378,1401):
    df_list.append(pd.read_excel(path.format(i)))
for i, df in enumerate(df_list):
    df=df.dropna(subset=['نماد'])
    df[['year', 'month', 'day']] = df['سال مالی'].str.split('/', expand=True)
    df = df.astype({"year":int}, errors='raise') 
    df = df.astype({"month":int}, errors='raise') 
    df = df.astype({"day":int}, errors='raise')
    df['سال']=df['year']
    msk = df['month'].le(5)
    df.loc[msk,'سال'] = df.loc[msk,'سال'] -1 
    df.rename(columns={"سال مالی": "date"}, inplace = True)
    df.rename(columns={"سال": "fiscal_year"}, inplace = True)
    df.rename(columns={"شرکت": "name"}, inplace = True)
    df.rename(columns={"بازار": "market"}, inplace = True)
    df.rename(columns={"نماد": "firm"}, inplace = True)
    df = df[['name','market','firm','date','fiscal_year','تلفیقی','حسابرسی شده','تجدید ارائه شده','وجوه نقد و موجودی‌های نزد بانک‌ها','سرمایه گذاری‌های کوتاه مدت','حسابها و اسناد دریافتنی','سایر حسابها و اسناد دریافتنی','موجودی مواد و کالا','پیش پرداخت‌ها','جمع دارایی‌های جاری','مطالبات بلند مدت','سرمایه گذاری‌های بلند مدت','دارایی‌های نامشهود','سایر دارایی‌ها','خالص دارایی‌های ثابت','جمع دارایی‌های غیر جاری','جمع کل دارایی‌ها','حسابها و اسناد پرداختنی تجاری','سایر حساب‌ها و اسناد پرداختنی','سپرده‏ها و پیش دریافت‌ها','مالیات پرداختنی','تسهیلات جاری مالی دریافتی','سود پرداختنی به سپرده گذاران','جمع بدهی‌های جاری','حسابها و اسناد پرداختنی بلند مدت','پیش دریافت های بلند مدت','تسهیلات مالی دریافتی بلند مدت','ذخیره مزایای پایان خدمت کارکنان','سایر بدهی‌ها','جمع بدهی‌های غیر جاری','جمع کل بدهی‌ها','سرمایه','سود و زیان انباشته','اندوخته قانونی','اندوخته احتیاطی','سایر اندوخته‌ها','جمع حقوق صاحبان سهام در پایان سال مالی','جمع کل بدهی‌ها و حقوق صاحبان سهام']]
    path2 = f"D:\\User\\Desktop\\thesis\\data\\tarazname\\balance sheet\\cleanshode\\{i+1378}.xlsx"
    df.to_excel(path2, index=False)
```

# Merging

```python
import glob
import pandas as pd
path = r"D:\User\Desktop\thesis\data\tarazname\balance sheet\cleanshode"
file_list = glob.glob(path + "/*.xlsx")
excl_list = []
for file in file_list:
    excl_list.append(pd.read_excel(file, dtype='unicode', engine='openpyxl'))
excl_merged = pd.DataFrame()
for excl_file in excl_list:
    excl_merged = excl_merged.append(excl_file, ignore_index=True)
excl_merged.to_excel(r'D:\User\Desktop\thesis\data\tarazname\balance sheet\cleanshode\MERGE.xlsx', index=False)
```
