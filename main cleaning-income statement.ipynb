```python
import pandas as pd
from pathlib import Path, PurePath
path = r'D:\User\Desktop\thesis\data\tarazname\income statement\Income Statement {}.xlsx'
df_list = []
for i in range(1377,1401):
    df_list.append(pd.read_excel(path.format(i)))
for i, df in enumerate(df_list):
    df=df.dropna(subset=['نماد'])
    df[['year', 'month', 'day']] = df['سال مالی'].str.split('/', expand=True)
    df = df.astype({"year":int}, errors='raise') 
    df = df.astype({"month":int}, errors='raise') 
    df = df.astype({"day":int}, errors='raise') 
    df['سال']=df['year']
    msk = df['month'].le(5)
    df.loc[msk,'سال'] = df.loc[msk,'سال'] -1 
    df.rename(columns={"سال مالی": "date"}, inplace = True)
    df.rename(columns={"سال": "fiscal_year"}, inplace = True)
    df.rename(columns={"شرکت": "name"}, inplace = True)
    df.rename(columns={"بازار": "market"}, inplace = True)
    df.rename(columns={"نماد": "firm"}, inplace = True)
    df=df.drop(columns=['year', 'month','day'])
    path2 = f"D:\\User\\Desktop\\thesis\\data\\tarazname\\income statement\\cleanshode\\{i+1377}.xlsx"
    df.to_excel(path2, index=False)
```

```python
import glob
import pandas as pd
path = r"D:\User\Desktop\thesis\data\tarazname\income statement\cleanshode"
file_list = glob.glob(path + "/*.xlsx")
excl_list = []
for file in file_list:
    excl_list.append(pd.read_excel(file, dtype='unicode', engine='openpyxl'))
excl_merged = pd.DataFrame()
for excl_file in excl_list:
    excl_merged = excl_merged.append(excl_file, ignore_index=True)
excl_merged.to_excel(r'D:\User\Desktop\thesis\data\tarazname\income statement\cleanshode\MERGE.xlsx', index=False)
```
